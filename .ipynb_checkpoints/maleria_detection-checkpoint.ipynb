```python
image_size=(224,224)
batch_size=32
train_datagen=ImageDataGenerator(preprocessing_function=preprocess_input,
                                 horizontal_flip=True,
                                 vertical_flip=True)
test_datagen=ImageDataGenerator(preprocessing_function=preprocess_input,
                                horizontal_flip=True,
                                vertical_flip=True)
```

```python
train_gen=train_datagen.flow_from_directory(train_set,
                                            target_size=image_size,
                                          
                                            )

valid_gen=test_datagen.flow_from_directory('/content/gdrive/MyDrive/malaria_detection/valid',
                                          target_size=image_size,
                                          batch_size=batch_size,)
```

```python
from tensorflow.keras.applications.vgg16 import VGG16
base_model = VGG16(input_shape=(224, 224, 3),
                  include_top = False,
                  weights = 'imagenet')
```

```python

for layer in base_model.layers:
    layer.trainable = False
```

```python
x = Flatten()(base_model.output)
x = Dense(512, activation='relu')(x)
x = Dense(2, activation='sigmoid')(x)
model = Model(base_model.input, x)
```

```python
patience = 1
stop_patience = 2
factor = 0.5

callbacks = [
    ModelCheckpoint("classify_model.h5", save_best_only=True, verbose = 0),
    EarlyStopping(patience=stop_patience, monitor='val_loss', verbose=1,l;m),
    ReduceLROnPlateau(monitor='val_loss', factor=factor, patience=patience, verbose=1)]
```

```python
epochs = 30
model.compile(optimizer = 'adam' , loss = 'binary_crossentropy', metrics = ['accuracy'])

history=model.fit(train_gen, validation_data=valid_gen, epochs=epochs,batch_size=64, callbacks=callbacks, verbose=1)
```

```python
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(valid_gen)
```
